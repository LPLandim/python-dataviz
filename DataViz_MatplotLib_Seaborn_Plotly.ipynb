# Dataset com Dados de Imigrantes para o Canadá

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv")

df
```

```python
df.info()
```

Mudando o index para o nome do país (facilita a busca)

```python
df.set_index('País', inplace= True)
```

Armazenando todas as colunas de tempo em uma variável

```python
anos = list(map(str, range(1980, 2014)))

anos
```

## Vamos focar no Brasil para essa análise de tendências

```python
brasil = df.loc['Brasil', anos]

brasil
```

Transformando em DataFrame

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
# A visão de anos não está muito boa, vamos melhorar
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) 
```

```python
plt.figure(figsize= (8, 4)) # Ajusta o tamanho do gráfico
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá') # Adiciona título ao gráfico


# Cria a lista de anos e imigrantes com um loop for
anos, imigrantes = [], []

for ano in range(1980, 2016, 5):  # Note que o final é 2016 para incluir o 2015
    anos.append(str(ano))

for imigrante in range(500, 3001, 500): # Note que o final é 3001 para incluir o 3000
    imigrantes.append(imigrante)  


plt.xlabel('Ano') # Dá nome ao eixo X
plt.ylabel('Número de imigrantes') # Dá nome ao eixo Y

plt.xticks(anos)  # Passa a lista para plt.xticks
plt.yticks(imigrantes) # Passa a lista para plt.yticks
plt.show()
```

# Comparando os números de imigrantes do Brasil e Argentina para o Canadá

## Que são os maiores países da América do Sul.

```python
anos = list(map(str, range(1980, 2014)))
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacao = df_comparacao.T # T de transpose (trocar linhas por colunas)

# Como resultado temos uma coluna com os anos e duas colunas para cada país contendo o número de imigrantes.

df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'], label = 'Brasil')
plt.plot(df_comparacao['Argentina'], label ='Argentina')

plt.title('Imigração do Brasil e Argentina para o Canadá')

plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

plt.legend()
plt.show()
```

Com este gráfico, podemos notar que as tendências de imigração no `Brasil` e na `Argentina` eram bastante semelhantes no início do período. No entanto, a partir dos anos 2000, os números de imigrantes na Argentina começaram a aumentar e atingiram os valores mais elevados entre 2000 e 2005, antes de cair e atingir valores muito baixos nos anos mais recentes. Por outro lado, no caso do Brasil, os aumentos na imigração começaram após 2005 e parecem ter atingido o valor mais alto em 2010.

Esses gráficos estão em escalas diferentes de eixo y. No gráfico do Brasil, o valor vai até `2500`, nos da Argentina e do Peru, até `1500`, e no da Colômbia, até `6000`. Isso pode gerar uma `confusão`

Vamos consertar.

```python
fig, axs = plt.subplots(2, 2, figsize = (10, 6))

# Ajustando as sobreposições dos gráficos
fig.subplots_adjust(hspace = 0.5, # Espaçamento vertical
                     wspace = 0.3) # Espaçamento horizontal
fig.suptitle("Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013")

# Primeiro gráfico
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

# Segundo gráfico
axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

# Terceiro gráfico
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

# Quarto gráfico
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat: # .flat faz um loop pra percorrer todos os subplots
     ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # type: ignore

for ax in axs.flat:
     ax.set_xlabel('Ano')
     ax.set_ylabel('Número de Imigrantes')

# Definindo os limites do eixo y
ymin = 0 
ymax = 7000

for ax in axs.ravel(): # .ravel transforma ou "achata" um array bidimensional num array unidimensional
     ax.set_ylim(ymin, ymax)
     

plt.show()
```

### Adicionando anotações nas barras para melhorar a visualização dos valores

```python
# Lista vazia de cores
cores = [] 

 # Loop for para iterar sobre todos os índices de países e quando for o Brasil, colocar uma cor específica
for pais in america_do_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')
     

fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_do_sul_sorted.index, america_do_sul_sorted['Total'], color = cores )

ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes para o Canadá\n1980 a 2013'
             , fontsize = 18) 
ax.set_xlabel('Número de imigrantes')
ax.set_ylabel('', fontsize = 14)

ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

# Loop for para percorrer sobre os valores na coluna "Total".
# função enumerate() é usada para obter tanto o índice (i) quanto o valor (v) de cada elemento na coluna "Total".

# v + 20: a posição horizontal do texto, que é o valor da coluna "Total" mais 20 unidades
 #                              (para afastar um pouco do final da barra).

# i: a posição vertical do texto, que é o índice do loop (correspondente a cada país).
# str(v): o texto a ser exibido, que é a conversão do valor da coluna "Total" para string.
# color='black': cor do texto
# fontsize=10: o tamanho da fonte do texto.
# ha='left': o alinhamento horizontal do texto, que é à esquerda.
# va='center': o alinhamento vertical do texto, que é centralizado.


for i, v in enumerate(america_do_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color = 'black', fontsize = 10, ha = 'left', va = 'center')

ax.set_frame_on(False) # Remove a caixa (frame) que envolve o gráfico
ax.get_xaxis().set_visible(False) # Oculta o eixo x (horizontal) do gráfico.

# modifica os parâmetros dos "ticks" (marcações) nos eixos do gráfico.
ax.tick_params(axis = 'both', which = 'both', length = 5) 
# axis='both': aplica as modificações tanto no eixo x quanto no eixo y.
# which='both': aplica as modificações tanto aos ticks principais quanto aos secundários.
# length=5: define o comprimento dos ticks como 5 unidades.

plt.grid(False)
plt.show()
```

### Mudando a paleta de cores

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize = (8, 4))
    ax = sns.barplot(data = top_10, y = top_10.index, x = 'Total', orient = 'h', palette = palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc = 'left', fontsize = 18)
    ax.set_xlabel('Número de imigrantes', fontsize =14)
    ax.set_ylabel('')

    plt.show()
```

### Paleta categórica

No entanto, o objetivo atual é exibir um "top 10" dos países com mais imigrantes, o que não se encaixa perfeitamente com essa paleta de `coolwarm`. Portanto, não seria a escolha mais adequada para o nosso caso.

Uma opção interessante a ser considerada é o uso de uma paleta categórica. Essas paletas são adequadas quando temos diferentes categorias, como é o caso dos países neste contexto.

Ao utilizar uma paleta categórica, evitamos que uma cor se sobressaia em relação às outras, o que é benéfico para destacar as informações distintas entre as categorias.

```python
gerar_grafico_paleta("tab10")
```
